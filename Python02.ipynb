* Operators in python

1. Arithmetic operators
2. Relational operators
3. Logical operators
4. Bitwise operators
5. Assignment operators
6. Membership operator

```python
# Arithmetric operators

a = 10
b = 20

print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a % b)  # remainder deta hai
print(a // b)  # integer division
print(a ** b)  # power of the operators
```

```python
# Relational operators

print(4>5)
print(4<5)
print(4>=4)
print(4<=4)
print(4==4)
print(4!=4)
```

```python
# Logical operators ---> AND , OR , NOT <----
print (1 and 0 )
print (1 or 0 )
print ( not 1 )
```

```python
# Bitwise operators

# bitwise and
print (2 & 3 )

# bitwise or
print (2 | 3)

# bitwise xor
print (2 ^ 3)

print (~3)

print(4 >> 2)
print (5 << 2)
```

```python
# Assignment operators

# =
# a = 2

a = 2
# a = a + 2
a  += 2
print(a)

```

```python
# Membership operators

# in / not in

print ('D' in 'Delhi')
print ('D' not in 'Delhi' )
```

```python
print (1 in [2,3,4,5,6])
```
