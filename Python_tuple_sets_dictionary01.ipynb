# Difference between Lists and Tuples

* Syntax
* Mutability
* Speed ---> immutable data is faster then list
* Memory ---memory kam leta hai tuple mein
* Built in functionality ---> functionlaity jaada hai list mein
aur tuple mein kam hai

* Error prone
* Usability

```python
# speed kam leta hai tuple me
import time

L = list(range(100000000))
T = tuple(range(100000000))

start = time.time()
for i in L:
    i*5
print('List time',time.time()-start )


start = time.time()
for i in T:
    i*5
print('Tuple time', time.time()-start)
```

```python
# memory kam leta hai tuple me

import sys

L = list(range(1000))
T = tuple(range(1000))

print('list size', sys.getsizeof(L))
print('tuple size ', sys.getsizeof(T))
```

# Sets

A set is an unordered collection of items. Every set element is unique( no duplicates) and must be immutable (cannot be changed ).

However, a set itself is mutable . we can add or remove items from it

sets can also be used to perform mathematicals like union , intersection , symmetric difference etc.


# Characterstics :

* Unordered
* Mutable
* No Duplicates
* Can't contain mutable data type
