```python
# reading from files
#  ----> using read()
# ----> file ko read kr sakte kaise python me jaise read karte hai
# bur strings formated me hota hai

f = open('/content/sample.txt' , 'r')
s = f.read()
print(s)
f.close()
```

```python
# reading upto n chars
# only read 10 character bas

f = open('/content/sample.txt' , 'r')
s = f.read( 10 )
print(s)
f.close()
```

```python
# readline() ---> to read line by line
# isme line change hota hai print karne ke baad automatic line change ho jata hai

f = open('/content/sample.txt' , 'r')
print(f.readline())
print(f.readline())
f.close()
```

```python
f = open('/content/sample.txt' , 'r')
print(f.readline(),end='')
print(f.readline(),end='')
f.close()
```

```python
# reading entire using readline

f = open('/content/sample.txt' , 'r')

while True:

  data= f.readline()
  if data == '':
    break
  else:
    print(data)
f.close()
```

```python
# reading entire using readline
# line change nhi hoti hai end='' this is a syntax

f = open('/content/sample.txt' , 'r')

while True:

  data= f.readline()
  if data == '':
    break
  else:
    print(data,end='')
f.close()
```

# Using Context Manager (With)

* it's a good idea to close a file after usage as it will free up the resources

* if we dont close it, garbage collector would close it

* with keyword closes the file as soon as the usage it over

```python
# With keyword
with open ('/content/sample1.txt' ,'w') as f:
  f.write('hello archi')
```

```python
f.write('hello')
```

```python
# try f.read() now

with open('/content/sample.txt' , 'r') as f:
  print(f.read())
```

```python
# moving within a file ---> 10 char then 10 char
with open('/content/sample.txt' , 'r') as f:
  print(f.read(10))
```

```python
with open('/content/sample.txt' , 'r') as f:
  print(f.read(10))
  print(f.read(10))
```

```python
# benefit ? ---> to load a big file in memory

big_L = ['hello world ' for i in range (1000)]

with open('big.text', 'w') as f :
  f.writelines(big_L)
```

```python
# benefit ? ---> to load a big file in memory


with open('big.text','r') as f:

  chunk_size = 100

  while len(f.read(chunk_size)) > 0:
    print(f.read(chunk_size), end='')
    f.read(chunk_size)
```
