# Sequence sum

1/1!+2/2!+3/3!+....

```python
# code here

n = int(input("Enter a number: "))

result = 0
fact = 1
for i in range(1, n+1):
    fact = fact * i
    result = result + i/fact
print(result)
```

# Pattern 1



```python
rows = int(input("Enter the number of rows: "))

for i in range(1,rows+1):
    for j in range(1,i+1):
        print('*',end='')
    print()
```

# Patterns 2

```python
rows = int(input("Enter the number of rows: "))

for i in range(1,rows+1):
    for j in range(1,i+1):
        print(j,end='')
    print()
```

```python
rows = int(input("Enter the number of rows: "))

for i in range(1,rows+1):
    for j in range(1,i+1):
        print(j,end='')
    for k in range (i-1,0,-1):
      print (k,end='')
    print()
```

# Loop control statement

* Break
* Continue
* Pass

```python
for i in range(1,10):
    print(i)
```

```python
for i in range(1,10):
    if i == 5:
        break
    print(i)
```

```python
# prime number

lower = int(input("Enter the lower range: "))
upper = int(input("Enter the upper range: "))

for i in range(lower,upper+1):
    for j in range(2,i):
      if i%j == 0 :
        break

    else :
      print (i)
```

```python
# continue

for i in range(1,10):
    if i == 5:
        continue
    print(i)
```

```python
for i in range(1,10):
    pass
```
