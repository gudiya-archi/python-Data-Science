# Creating a list

```python
# Empty
print([])

# 1D    ---> homogenous
print([1,2,3,4])

# 2D
print([1,2,3,[4,5,6]])  # nested list , hetrogenous
# 3D
print([[[1,2],[3,4]],[[5,6],[7,8]]]) # homo

# Hetrogenous
print([1, True,5.6,5+6j,'hello'])

# using type conversion
print(list('hello'))
```

# Accessing items from a list

```python
# Indexing
L = [1,2,3,4,5]
print(L[0])   # positive
#             ----> left to right jaata hai aur 0 se started hota hai
```

```python
L = [1,2,3,4,5]
print(L[2])
```

```python
#  negative indexing
# right to left rahta hai
# aur -1 se started hota hai

L = [1,2,3,4,5]
print(L[-1])
```

```python
L = [1,2,3,[4,5]]
print(L[-1])   # 2D list
```

```python
L = [1,2,3,[4,5]]
print(L[-1][-2])   # 2D list
```

```python
L = [1,2,3,[4,5]]
print(L[3][0])
```

```python
L = ([[[1,2],[3,4]],[[5,6],[7,8]]])
print(L[0] )
```

# Adding items to a list

* append
* extend
* insert

```python
# append
# -----> append are used only one items in the list

L = [1,2,3,4,5]
L.append(6)
print(L)
```

```python
# append
L = [1,2,3,4,5]
L.append(True)
print(L)
```

```python
# extend
# -----> extend are used multipy item in the list
L = [1,2,3,4,5]
L.extend([6,7,8])
print(L)
```

```python
 # append
L = [1,2,3,4,5]
L.append([6,7,8])  # isko list maine dal diya hai
print(L)
```

```python
# extend
L = [1,2,3,4,5]
L.extend('delhi')
print(L)
```

```python
# insert
L = [1,2,3,4,5]
L.insert(1, 100)
print(L)
```

```python
s = 'hello'
s[0] = 'x' # string are immutable
print(s)
```

# Editing items in a list

```python
# list maine change hota hai ..... mutable hai
L = [1,2,3,4,5]
L[-1] = 100
print(L)
```

```python
L = [1,2,3,4,5]
# editing with indexing
L[-1] = 500

# editing with slicing
L[1:4] = [200, 300, 400]
print(L)
```
